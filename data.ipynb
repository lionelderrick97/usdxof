```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns

# Importing the dataset
df = pd.read_csv('usdxof.csv')
df
```

```python
# Extraire l'année de chaque date
data['annee'] = data['Date'].dt.year
# Regrouper les données par année et calculer la variation de "price"
variations = data.groupby('annee')['Price'].apply(lambda x: (x.iloc[-1] - x.iloc[0]) / x.iloc[0] * 100)

# Tracer un graphique en barres de la variation de "price" pour chaque année
variations.plot(kind='bar')
plt.title('Variation de Price pour chaque année')
plt.xlabel('Année')
plt.ylabel('% de variation')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Grouper par année et calculer la moyenne de la colonne 'price'
data_annee_price = data.groupby('annee')['Price'].mean().reset_index()

# Initialiser la figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Créer le graphique 3D avec les données de la variation du prix par année
ax.scatter(data_annee_price['annee'], [0]*data_annee_price.shape[0], data_annee_price['Price'])

# Afficher le graphique
plt.show()
```
